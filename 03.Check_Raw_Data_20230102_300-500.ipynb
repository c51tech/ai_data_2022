```python
def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for i, f in enumerate(files):
            if i >= 3:
                print('{}...'.format(subindent))
                break
                
            print('{}{}'.format(subindent, f))
            
```

### 데이터 폴더 확인

```python
def sort_dirs_by_specimen_id(root_dir):
    specimen_dirs = os.listdir(root_dir)
    specimen_ids = [d.split('_')[0] for d in specimen_dirs]
    specimen_ids = [int(i) for i in specimen_ids if i.isdigit()]
    
    return [d for _, d in sorted(zip(specimen_ids, specimen_dirs))]
```

```python
err_cnt = 0
specimen_dirs = sort_dirs_by_specimen_id(image_root_dir)
complete_specimen_id_list = []

for specimen_id in sorted(specimen_dirs):
    
    try:
        assert os.path.exists(os.path.join(image_root_dir, specimen_id, 'Mechnical property')), specimen_id
        assert os.path.exists(os.path.join(image_root_dir, specimen_id, 'Off_Axis_Images')), specimen_id
        assert os.path.exists(os.path.join(image_root_dir, specimen_id, 'On_Axis_Images')), specimen_id
#         assert os.path.exists(os.path.join(image_root_dir, specimen_id, 'melt_temperature')), specimen_id
        assert os.path.exists(os.path.join(image_root_dir, specimen_id, 'video')), specimen_id
#         assert os.path.exists(os.path.join(image_root_dir, specimen_id, 'Tensile_Images')), specimen_id

        complete_specimen_id_list.append(specimen_id)
    except:
        err_cnt += 1
        print(specimen_id, os.listdir(os.path.join(image_root_dir, specimen_id)))

print(f'{len(specimen_dirs) - err_cnt} / {len(specimen_dirs)} specimen has passed!')
```

#### JSON 구문 규칙 불러오기

```python
df_img_desc = pd.read_excel('data_meta_info/기계적 특성치 데이터 이미지별_rule_2022-12-26.xlsx', header=1)
df_img_desc.columns = ['항목'] + df_img_desc.columns[1:].tolist()
df_img_desc
```

```python
json_rule_dict = {}
for i, row in df_img_desc.iterrows():
    name = row['항목']
    
    _row_dict = row.iloc[1:].to_dict()
    if isinstance(_row_dict['유효값'], str):
        _row_dict['유효값'] = [s.replace('"', '').replace('\n', '') for s in _row_dict['유효값'].split(',')]
    
    json_rule_dict[name] = _row_dict
    
json_rule_dict
```

#### 구문검사 실행

```python
regex_num_unit = re.compile(r'([0-9\.]+)(.*)')

for json_file_name in on_axis_json_list:
    with open(json_file_name) as f:
        _json_dict = json.load(f)
    
    ## 필수항목 포함 여부 확인
    for k, v in json_rule_dict.items():
        if k not in _json_dict:
            if v['필수여부'] == 'y':
                print(json_file_name, f"필수 항목 누락: {k}")

    for k, v in _json_dict.items():
        ## 미등록 항목 포함 여부 확인
        if k not in json_rule_dict:
            print(json_file_name, f"항목 오류: {k}")
            continue

        ## 타입 체크(모든 항목의 타입이 string임)
        if json_rule_dict[k]['타입'] == 'string' and not isinstance(v, str):
            print(json_file_name, f"타입 오류: {k}", f"{json_rule_dict[k]['타입']}, but {type(v)}")

        ## 유효값 체크
        if isinstance(json_rule_dict[k]['유효값'], list) and v not in json_rule_dict[k]['유효값']:
            print(json_file_name, f"유효값 오류: {k}", f"{json_rule_dict[k]['유효값']}, but {v}")

        ## 최소값 체크
        if not np.isnan(json_rule_dict[k]['최소값']):
            m = regex_num_unit.match(v)
            if m is None:
                print(json_file_name, f"최소값 오류: {k}", f"숫자+단위 형식이 아님({v})")
            elif float(m[1]) < json_rule_dict[k]['최소값']:
                print(json_file_name, f"최소값 오류: {k}", f"{v} > {json_rule_dict[k]['최소값']}")

        ## 최대값 체크
        if not np.isnan(json_rule_dict[k]['최대값']):
            m = regex_num_unit.match(v)
            if m is None:
                print(json_file_name, f"최대값 오류: {k}", f"숫자+단위 형식이 아님({v})")
            elif float(m[1]) > json_rule_dict[k]['최대값']:
                print(json_file_name, f"최대값 오류: {k}", f"{v} < {json_rule_dict[k]['최대값']}")

        ## null 체크
        if v is None and not (isinstance(json_rule_dict[k]['null타입허용여부'], str) 
                                and json_rule_dict[k]['null타입허용여부'] == 'y'):
            print(json_file_name, f"null 오류: {k}")
            
print("구문 검사 종료")
```

```python
## 기계적 특성치가 0이 아닌지 확인
mech_prop_keys = ['yielding.stress', 'tensile.stress', 'elongation', 'density']

for _json_path in on_axis_json_list:
    
    with open(_json_path) as f:
        json_dict = json.load(f)
        
    for k in mech_prop_keys:
        if k not in json_dict:
            print(_json_path, f'특성치 누락: {k}')
            continue
        
        try:
            v = json_dict[k]
            m = regex_num_unit.match(v)
            value = float(m.groups()[0])
            
            assert value > 0.0
        except Exception as e:
            err_cnt += 1
            print(_json_path, e, k, f'[{v}]')
            
print("특성치 검사 종료")
    
```
