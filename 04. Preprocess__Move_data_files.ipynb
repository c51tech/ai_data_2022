# 200 장비 데이터 정리 및 이동

```python
## 200 장비 데이터 폴더
data_root = '/home/kikim/data/ai_data_230102/200/AI DB 금속 스파크 이미지 데이터'
image_root_dir = os.path.join(data_root, '원천데이터')
label_root_dir = os.path.join(data_root, '라벨데이터')
```

```python
## 동영상 데이터 파일 리스트
video_file_list = []
valid_specimen_dir_list = []

for d in os.listdir(image_root_dir):
    if not d.split('_')[0].isdigit():
        print(f"'{d}' is not specimen folder. Ignore '{d}'")
        continue
        
    _video_folder = os.path.join(image_root_dir, d, 'video')
    _video_files = [os.path.join(_video_folder, f) for f in os.listdir(_video_folder) if f.lower().endswith('.avi')]
    
    if len(_video_files) > 0:
        valid_specimen_dir_list.append(d)
        video_file_list.extend(_video_files)
```

```python
print(len(valid_specimen_dir_list))

display(valid_specimen_dir_list[:5])
```

```python
print(len(video_file_list))

display(video_file_list[:5])
```

```python
## 메타 데이터 파일 리스트

on_axis_json_file_list = []
off_axis_json_file_list = []

err_specimen_dir_list = []
for d in valid_specimen_dir_list:
    _on_axis_dir = os.path.join(label_root_dir, d, 'On_Axis_Images', 'JSON')
    _off_axis_dir = os.path.join(label_root_dir, d, 'Off_Axis_Images', 'JSON')
    
    try:
        _1st_on_axis_json_file = [f for f in os.listdir(_on_axis_dir) if f.endswith('.json')][0]
        _1st_off_axis_json_file = [f for f in os.listdir(_off_axis_dir) if f.endswith('.json')][0]

        on_axis_json_file_list.append(os.path.join(_on_axis_dir, _1st_on_axis_json_file))
        off_axis_json_file_list.append(os.path.join(_off_axis_dir, _1st_off_axis_json_file))
    except Exception as e:
        print(d, e)
        err_specimen_dir_list.append(d)
```

```python
## 에러가 발생한 폴더 제외

valid_specimen_dir_list = [d for d in valid_specimen_dir_list if d not in err_specimen_dir_list]
video_file_list = [f for f in video_file_list for s in err_specimen_dir_list if not s in f]

print(len(valid_specimen_dir_list))
print(len(video_file_list))
```

## 데이터 카피

```python
## On_Axis Json 파일 카피
for s_path in on_axis_json_file_list:
    d_dir = os.path.join(data_dest, '라벨데이터', s_path.split('/')[-4], 'On_Axis_Images', 'JSON')
    os.makedirs(d_dir, exist_ok=True)
    shutil.copy(s_path, d_dir)
```

```python
## Off_Axis Json 파일 카피
for s_path in off_axis_json_file_list:
    d_dir = os.path.join(data_dest, '라벨데이터', s_path.split('/')[-4], 'Off_Axis_Images', 'JSON')
    os.makedirs(d_dir, exist_ok=True)
    shutil.copy(s_path, d_dir)
```

```python
## video 파일 카피
for s_path in video_file_list:
    d_dir = os.path.join(data_dest, '원천데이터', s_path.split('/')[-3], 'video')
    os.makedirs(d_dir, exist_ok=True)
    shutil.copy(s_path, d_dir)
```

# 300, 500 장비 데이터 정리 및 이동

```python
## 200 장비 데이터 폴더
data_root = '/home/kikim/data/ai_data_230102/300_500/AI DB 금속 스파크 이미지 데이터'
image_root_dir = os.path.join(data_root, '원천데이터', "기계적 특성치 데이터")
label_root_dir = os.path.join(data_root, '라벨링데이터', "기계적 특성치 데이터")
```

```python
## 동영상 데이터 파일 리스트
video_file_list = []
valid_specimen_dir_list = []

for d in os.listdir(image_root_dir):
    if not d.split('_')[0].isdigit():
        print(f"'{d}' is not specimen folder. Ignore '{d}'")
        continue
        
    _video_folder = os.path.join(image_root_dir, d, 'video')
    _video_files = [os.path.join(_video_folder, f) for f in os.listdir(_video_folder) if f.lower().endswith('.avi')]
    
    if len(_video_files) > 0:
        valid_specimen_dir_list.append(d)
        video_file_list.extend(_video_files)
```

```python
print(len(valid_specimen_dir_list))

display(valid_specimen_dir_list[:5])
```

```python
print(len(video_file_list))

display(video_file_list[:5])
```

```python
## 메타 데이터 파일 리스트

on_axis_json_file_list = []
off_axis_json_file_list = []

err_specimen_dir_list = []
for d in valid_specimen_dir_list:
    _on_axis_dir = os.path.join(label_root_dir, d, 'On_Axis_Images', 'JSON')
    _off_axis_dir = os.path.join(label_root_dir, d, 'Off_Axis_Images', 'JSON')
    
    try:
        _1st_on_axis_json_file = [f for f in os.listdir(_on_axis_dir) if f.endswith('.json')][0]
        _1st_off_axis_json_file = [f for f in os.listdir(_off_axis_dir) if f.endswith('.json')][0]

        on_axis_json_file_list.append(os.path.join(_on_axis_dir, _1st_on_axis_json_file))
        off_axis_json_file_list.append(os.path.join(_off_axis_dir, _1st_off_axis_json_file))
    except Exception as e:
        print(d, e)
        err_specimen_dir_list.append(d)
```

```python
## 온도 데이터 파일 리스트
melt_temp_json_dir_list = []

for d in valid_specimen_dir_list:
    _melt_temp_dir = os.path.join(label_root_dir, d, 'melt_temperature')
    if os.path.exists(_melt_temp_dir) and len([f for f in os.listdir(_melt_temp_dir) if f.endswith('.json')]) > 0:
        melt_temp_json_dir_list.append(_melt_temp_dir)
```

```python
len(melt_temp_json_dir_list)
```

```python
## 에러가 발생한 폴더 제외
if len(err_specimen_dir_list) > 0:
    valid_specimen_dir_list = [d for d in valid_specimen_dir_list if d not in err_specimen_dir_list]
    video_file_list = [f for f in video_file_list for s in err_specimen_dir_list if not s in f]

print(len(valid_specimen_dir_list))
print(len(video_file_list))
```

## 데이터 카피

```python
## On_Axis Json 파일 카피
for s_path in on_axis_json_file_list:
    d_dir = os.path.join(data_dest, '라벨데이터', s_path.split('/')[-4], 'On_Axis_Images', 'JSON')
    os.makedirs(d_dir, exist_ok=True)
    shutil.copy(s_path, d_dir)
```

```python
## Off_Axis Json 파일 카피
for s_path in off_axis_json_file_list:
    d_dir = os.path.join(data_dest, '라벨데이터', s_path.split('/')[-4], 'Off_Axis_Images', 'JSON')
    os.makedirs(d_dir, exist_ok=True)
    shutil.copy(s_path, d_dir)
```

```python
## video 파일 카피
for s_path in video_file_list:
    d_dir = os.path.join(data_dest, '원천데이터', s_path.split('/')[-3], 'video')
    os.makedirs(d_dir, exist_ok=True)
    shutil.copy(s_path, d_dir)
```

```python
## melt_temperature 폴더 카피
for s_path in melt_temp_json_dir_list:
    d_dir = os.path.join(data_dest, '라벨데이터', s_path.split('/')[-2], 'melt_temperature')
    shutil.copytree(s_path, d_dir)
```
